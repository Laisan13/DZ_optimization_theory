При фиксированном seed=42 поразбирайтесь и поэкспериментируйте с параметрами алгоритма дифференциальной эволюции: strategy, popsize, tol, mutation, recombination и updating.

Постройте графики количества итераций оптимизации функции ackley от значения параметра.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import arange
from numpy import meshgrid
from scipy.optimize import dual_annealing
from scipy.optimize import differential_evolution
from scipy.optimize import basinhopping
from matplotlib import pyplot
```

```python
def ackley (x):
    arg1 = -0.2 * np . sqrt( 0.5 * (x[ 0 ] ** 2 + x[ 1 ] ** 2 ))
    arg2 = 0.5 * (np . cos( 2. * np . pi * x[ 0 ]) + np . cos( 2. * np . pi * x[ 1 ]))
    return -20. * np . exp(arg1) - np . exp(arg2) + 20. + np . e

bounds = [( -5 , 5 ), ( -5 , 5 )]
#strategy управляет типом выполняемого дифференциального поиска эволюции. По умолчанию установлено значение « best1bin »
import pandas as pd
strat =('best1bin','best1exp','rand1exp', 'randtobest1exp','currenttobest1exp', 'best2exp', 'rand2exp', 'randtobest1bin','currenttobest1bin','best2bin', 'rand2bin','rand1bin')
df = pd.DataFrame()
for i in strat:
    result = differential_evolution(ackley, bounds, seed =42, strategy=i)
    new_values= pd.Series([result.nfev], index=[i])
    df= df.append(new_values, ignore_index=True)
    print(result.nfev,result.x)

df.plot(kind='bar')
plt.show()
```

```python
#popsize контролирует размер или количество возможных решений, поддерживаемых в популяции по умолчанию 15
nfev1 = []
popsize1 = []
for i in range(0, 30 ,1):
    result2 = differential_evolution(ackley, bounds, seed =42, popsize= i)
    nfev1.append(result2.nfev)
    popsize1.append(i)
    print(result2.nfev,result2.x,i )

plt.plot( nfev1, popsize1 )  
```

```python
#tol - Относительный допуск для схождения
nfev1 = []
tol1 = []
for i in np.arange(0.1,2,0.1):
    result3 = differential_evolution(ackley, bounds, seed =42,  tol = i)
    nfev1.append(result3.nfev)
    tol1.append(i)
    print(result3.nfev,result3.x,i )    

plt.plot( nfev1, tol1)
```

```python
#mutation - Константа мутации. В литературе это также известно как дифференциальный вес, обозначаемый буквой F  
#если указан как число с плавающей запятой, он должен находиться в диапазоне [0, 2]
nfev1 = []
mutation1 = []
for i in np.arange(0,2,0.1):
    result4 = differential_evolution(ackley, bounds, seed =42,  mutation = i)
    nfev1.append(result4.nfev)
    mutation1.append(i)
    print(result4.nfev,result4.x,i )
plt.plot( nfev1, mutation1)
```

```python
# recombination - Константа рекомбинации должна быть в диапазоне [0, 1]. В литературе это также известно как вероятность кроссовера  
#Увеличение этого значения позволяет большему количеству мутантов перейти в следующее поколение
nfev1 = []
recombination1 = []
for i in np.arange(0,1.1,0.1):
    result5 = differential_evolution(ackley, bounds, seed =42,   recombination = i)
    nfev1.append(result5.nfev)
    recombination1.append(i)
    print(result5.nfev,result5.x,i )

plt.plot( nfev1, recombination1)
```

```python
# updating- значения 'immediate' или 'deferred'
# Если 'immediate', вектор наилучшего решения постоянно обновляется в пределах одного поколения
# При наличии 'deferred'вектор наилучшего решения обновляется один раз за поколение
updat =('immediate','deferred')
df = pd.DataFrame()
for i in updat :
    result6 = differential_evolution(ackley, bounds, seed =42, updating = i)
    new_values= pd.Series([result6.nfev], index=[i])
    df= df.append(new_values, ignore_index=True)

df.plot(kind='bar')
plt.show()
```
